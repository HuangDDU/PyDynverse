```python
import pydynverse as pdv
```

```python
dataset = pdv.data.load_simulation_data(data_filename="synthetic/dyntoy/bifurcating_1.rds")
dataset
```

```python
# TODO: milestone颜色调整
pdv.plot.plot_graph(dataset)

trajectory_edges_group = pdv.wrap.group_onto_nearest_milestones(dataset) # 聚合到最近的milestone上
pdv.plot.plot_graph(dataset, grouping=trajectory_edges_group)

trajectory_edges_group = pdv.wrap.group_onto_trajectory_edges(dataset) # 聚合到轨迹边上
pdv.plot.plot_graph(dataset, grouping=trajectory_edges_group)
```
